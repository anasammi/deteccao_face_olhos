```python
#importar openCV
import cv2
from google.colab.patches import cv2_imshow
```

```python
#carregar imagem
image = cv2.imread('/content/rosto.jpg')
```

```python
#importar arquivo em cascata para reconhecimento facial
face_classifier = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
#converter imagem para escala de cinza
imgGray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(imgGray)
```

```python
#classificador retorna a face detectada como uma tupla
#armazena a coordenada superior esquerda e as coordenadas inferior direita
faces = face_classifier.detectMultiScale(imgGray, scaleFactor = 1.3, minNeighbors = 5)
```

```python
#Se faces dor vazio, significa que o algoritmo não encontrou nenhum rosto
if faces is ():
  print('Nenhum rosto encontrado')

#iterar sobre o array faces para desenhar um ret^ngulo sobre cada face em faces
for(x,y,w,h) in faces:
  cv2.rectangle(image, (x,y), (x+w, y+h), (127, 0,255), 2)

cv2_imshow(image)
```

```python
#detecção de olhos e face
olhosCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye_tree_eyeglasses.xml')

olhos = olhosCascade.detectMultiScale(imgGray, scaleFactor = 1.2, minNeighbors = 1)

if olhos is ():
  print("Nenhum olho encontrado")

for(x,y,w,h) in olhos:
  cv2.rectangle(image, (x,y), (x+w, y+h), (127, 0,255), 2)

cv2_imshow(image)
```
